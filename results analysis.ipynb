```python
import matplotlib.pyplot as plt
import numpy as np

# Data for VGG16
vgg16_train_loss = [0.9314, 0.7013, 0.5994, 0.5262, 0.4584, 0.3980, 0.3475, 0.2930, 0.2478, 0.2055]
vgg16_train_acc = [67.75, 75.45, 79.00, 81.71, 84.00, 86.11, 87.79, 89.79, 91.27, 92.72]

# Data for ResNet-50
resnet_train_loss = [0.6352, 0.6061, 0.5933, 0.5878, 0.5778, 0.5697, 0.5653, 0.5604, 0.5576, 0.5514]
resnet_train_acc = [77.65, 78.61, 79.09, 79.34, 79.85, 79.93, 80.16, 80.52, 80.55, 80.80]

# Data for AlexNet
alexnet_train_loss = [1.1704, 0.9077, 0.8195, 0.7574, 0.7174, 0.6749, 0.6435, 0.6165, 0.5892, 0.5639]
alexnet_train_acc = [59.76, 67.97, 71.24, 73.31, 74.70, 76.26, 77.11, 78.24, 79.09, 80.20]

# Plotting
plt.figure(figsize=(15, 10))

# Training Loss Plot
plt.subplot(2, 1, 1)
plt.plot(range(1, len(vgg16_train_loss) + 1), vgg16_train_loss, label='VGG16', marker='o')
plt.plot(range(1, len(resnet_train_loss) + 1), resnet_train_loss, label='ResNet-50', marker='s')
plt.plot(range(1, len(alexnet_train_loss) + 1), alexnet_train_loss, label='AlexNet', marker='^')
plt.title('Training Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.legend()
plt.grid(True)

# Training Accuracy Plot
plt.subplot(2, 1, 2)
plt.plot(range(1, len(vgg16_train_acc) + 1), vgg16_train_acc, label='VGG16', marker='o')
plt.plot(range(1, len(resnet_train_acc) + 1), resnet_train_acc, label='ResNet-50', marker='s')
plt.plot(range(1, len(alexnet_train_acc) + 1), alexnet_train_acc, label='AlexNet', marker='^')
plt.title('Training Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Training Accuracy (%)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
